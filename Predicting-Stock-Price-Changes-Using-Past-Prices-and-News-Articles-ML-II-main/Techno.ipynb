```python
import pandas as pd
import sklearn
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

dataset=pd.read_csv("/content/HistoricalData_1624010924150.csv",parse_dates=["Date"],index_col="Date")
```

```python
dataset.head(10)
```

# **`Checking for Missing Values`**

```python
missing = dataset.isnull().sum()
missing
```

# **Replacing $ with " "**

```python
dataset['Open']=dataset['Open'].str.replace('$',"")
dataset['Close/Last']=dataset['Close/Last'].str.replace('$',"")
dataset['High']=dataset['High'].str.replace('$',"")
dataset['Low']=dataset['Low'].str.replace('$',"")
dataset.head()
```

# **Converting values to float**

```python
dataset['Open'] = dataset['Open'].astype(float)
dataset['High'] = dataset['High'].astype(float)
dataset['Low'] = dataset['Low'].astype(float)
dataset['Close/Last'] = dataset['Close/Last'].astype(float)
dataset.head()
```

# **Data Visualization for Yearly values**

```python
plt.figure(figsize=(15,6))
sns.lineplot(data=dataset['Open'])
```

```python
plt.figure(figsize=(15,6))
sns.lineplot(data=dataset['Close/Last'])
```
